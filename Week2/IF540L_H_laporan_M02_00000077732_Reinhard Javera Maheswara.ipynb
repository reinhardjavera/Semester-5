```python
import pandas as pd
df= pd.DataFrame([['green', 'M', 10.1, 'class2'],
                  ['red', 'L', 13.5, 'class1'],
                  ['blue', 'XL', 15.3, 'class2']])

df.columns = ['color', 'size', 'price', 'classlabel']
```

```python
size_mapping = {'XL': 3,
                'L': 2,
                'M': 1}

df['size'] = df['size'].map(size_mapping)
```

```python
inv_size_mapping = {v: k for k, v in size_mapping.items()}
df['size'].map(inv_size_mapping)
```

```python
import numpy as np

#Create a mapping dict
#to convert class labels from strings to integers
class_mapping = {label: idx for idx, label in enumerate(np.unique(df['classlabel']))}
```

```python
#to convert class labels from string to integers
df['classlabel'] = df['classlabel'].map(class_mapping)
```

```python
#Reverse the class label mapping
inv_class_mapping = {v: k for k, v in class_mapping.items()}
df['classlabel'] = df['classlabel'].map(inv_class_mapping)
```

```python
from sklearn.preprocessing import LabelEncoder

#Label encoding with sklearn's LabelEncoder
class_le = LabelEncoder()
y = class_le.fit_transform(df['classlabel'].values)
y
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.preprocessing import StandardScaler
# 1. Load the Dataset
# Make sure to replace 'Steel_industry_data.csv' with the correct file path if necessary
df = pd.read_csv('E:/_Kuliah/Semester 5/Machine Learning/Week2/Steel_industry_data.csv')
# 2. Convert 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')
# 3. Extract useful features from the 'date' column
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
# 4. Drop the original 'date' column as it's now redundant
df.drop(columns=['date'], inplace=True)
# 5. Separate the target variable 'Load_Type'
y = df['Load_Type'] # Target variable
# 6. Drop 'Load_Type' from the DataFrame to prepare features
X = df.drop('Load_Type', axis=1)
# 7. Convert categorical variables to numeric using one-hot encoding
X = pd.get_dummies(X, columns=['WeekStatus', 'Day_of_week'], drop_first=True)
# 8. Convert target variable to numeric (Label Encoding)
y = y.map({'Light_Load': 0, 'Medium_Load': 1, 'Maximum_Load': 2})
# 9. Scale Numerical Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# 10. Split the Dataset into Training and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0, stratify=y)
# 11. Train a RandomForest Model to Determine Feature Importance
clf = RandomForestClassifier(n_estimators=100, random_state=0)
clf.fit(X_train, y_train)
# 12. Use the Trained Model to Select Important Features
model = SelectFromModel(clf, prefit=True)
X_train_selected = model.transform(X_train)
X_test_selected = model.transform(X_test)
# 13. Print Results
print("Original number of features:", X_train.shape[1])
print("Reduced number of features:", X_train_selected.shape[1])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
# Load the dataset
df = pd.read_csv('Steel_industry_data.csv')
```

```python
# Check for missing values
print("Missing values in each column:\n", df.isnull().sum())
# Impute missing values with the mean for numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
df[numerical_columns] = imputer.fit_transform(df[numerical_columns])
# Check for remaining missing values (should be zero)
print("Remaining missing values after imputation:\n", df.isnull().sum())
```

```python
# Convert 'date' column to datetime format and extract features
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
# Drop the original 'date' column
df.drop(columns=['date'], inplace=True)
```

```python
# Encode 'WeekStatus' and 'Day_of_week' using one-hot encoding
df = pd.get_dummies(df, columns=['WeekStatus', 'Day_of_week'], drop_first=True)
# Encode 'Load_Type' (target variable) using label encoding
class_le = LabelEncoder()
df['Load_Type'] = class_le.fit_transform(df['Load_Type'])
```

```python
# Separate features (X) and target (y)
X = df.drop('Load_Type', axis=1) # Features
y = df['Load_Type'] # Target variable
# Scale the features using Min-Max scaling
mms = MinMaxScaler()
X_norm = mms.fit_transform(X)
# Alternatively, standardize the features
stdsc = StandardScaler()
X_std = stdsc.fit_transform(X)
# Example to print normalized and standardized data
print("Normalized Data Sample:\n", X_norm[:5])
print("Standardized Data Sample:\n", X_std[:5])
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=0, stratify=y)
print("Training and testing data prepared.")
```
